## Computation and Problems

Given a question, __computation__ is the process we use to produce an answer.

The __computational problem__ is just that question which requires computation.

If that definition is too circular, we can use Oxford's definition "an inquiry starting from given conditions to investigate or demonstrate a fact, result, or law"

## Problems and Problem Instances

A problem instance is a particular problem with a particular solution.
- What are the prime factors of 15?
- Is 973624872469023461346013461387464763871273921731924791 prime?
- What is the shortest route to visit Austin, Dallas, Houston, and San Antonio?

A computational problem is a set of problem instances.
- What are the prime factors of a given integer?
- Is a given number prime?
- The travelling salesman problem

The size of a problem is defined as the __number of bits__ needed to describe an instance of the problem.

### Solving Computational Problems

Given a problem instance of size n, how long does it take to solve?
- Time could be substituted for other physical resources.
- For a polynomial time algorithm, the time scales as $n^{p}$ where p is the degree of the polynomial.
- For an exponential time algorithm, the time scales as $e^{r n}$ for r > 0.

The __computational complexity__ of a problem refers to how many resources are necessary to solve the problem.

### O-Notation - Describing the scaling of a problem

$\text{A function }g(n) \text{ is in }\mathcal{O}\big( f(n) \big) \text{ if there exist some constants }c,n_{0}\gt 0 \text{ such that for all }n\gt n_{0} \quad \frac{g(n)}{f(n)}\leq c $

For example: $$3n^{2} \in \mathcal{O}\big(n^{2}\big)$$ because $\frac{3n^{2}}{n^{2}}\leq 3$

#### Other O notations


$\Omega\big( f(n) \big) \text{ denotes functions }g(n)\text{ with }c,n_{0}\gt 0 \text{ such that for all }n\gt n_{0} \quad \frac{g(n)}{f(n)}\color{red}{\geq} c $

$\Theta\big( f(n) \big) \text{ denotes functions }g(n)\text{ such that } g(n) \in \mathcal{O}\big( f(n) \big) \text{ and } g(n) \in \Omega\big( f(n) \big)$

$o\big( f(n) \big) \text{ denotes functions }g(n)\text{ such that } \lim_{n\to\infty} \frac{g(n)}{f(n)}=0 $

$\omega\big( f(n) \big) \text{ denotes functions }g(n)\text{ such that } \lim_{n\to\infty} \frac{f(n)}{g(n)}=0 $

Note: people often say "of order" referring to how a function scales without being precise about exactly which O-notation they mean.

## Complexity Classes P, NP, BPP, BQP

P (Polynomial):
- The class of problems that can be solved in polynomial time
- Resources scale "polynomially" with the problem size \\
e.g. for a problem of size $n$ it might take $4n^3+n$ time to solve
- Considered “easy”

NP (Nondeterministic Polynomial):
- The class of problems whose solutions can be verified in polynomial time
- NP includes all the problems in P and maybe others which require superpolynomial resources to find solutions but only polynomial resources to check solutions.
- It is not known whether P = NP though it is widely suspected that this equality does not hold.

BPP (Bounded-error Probabilistic Polynomial):
- The class of problems that can be solved w/ probability 2/3 in polynomial time.  (Note that we need the probability bounded away from 1/2.)
- Note that this class allows for incorrect solutions to be found occasionally!

BQP (Bounded-error Quantum Polynomial):
- The class of problems that can be solved by a quantum computer w/ probability 2/3 in polynomial time.  (Note that “time” is related to number of qubits.)


### Other Important Complexity Classes

EXPTIME
- Problems that require exponentially increasing time to solve
- Examples: Unstructured Search, Checkers, Chess, Go, etc.

NP-hard
- At least as hard as the hardest problem in NP
- Require at most exponential time to solve

NP-complete
- Both NP and NP-hard
- Polynomially equivalent to any other NP-hard problem
- Examples: 3-SAT (Satisfiability), Traveling Salesman Problem, Super Mario Bros.


## Extended Church-Turing Thesis

The Church-Turing thesis essentially states that any physical process can be simulated with at Turing machine.
- A physical process can be said to be a type of computation
- The “extended” thesis restricts us to "efficient" simulations.
- Here “efficient” means with resources scaling at most polynomially.

Quantum Computational Supremacy
- It is unknown whether quantum computers violate the extended Church-Turing thesis
- Grover’s algorithm does not violate it.  (Still exponential scaling)
- Shor’s algorithm might violate it if factoring is actually hard.  (Not yet proven)
